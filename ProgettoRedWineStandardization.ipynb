```python
reviews = []
for i in data['quality']:
    if i >= 1 and i <= 6:
        reviews.append('0')
    elif i >= 7 and i <= 10:
        reviews.append('1')
data['Reviews'] = reviews
```

```python
data['Reviews'].unique()
```

```python
Counter(data['Reviews'])
```

```python
x = data.iloc[:,:11]
y = data['Reviews']
```

```python
x.head(10)
```

```python
y.head(10)
```

```python
from sklearn.preprocessing import StandardScaler
no = StandardScaler()
x = no.fit_transform(x)
print(x)
```

```python
from sklearn.decomposition import PCA
pca = PCA()
x_pca = pca.fit_transform(x)
```

```python
plt.figure(figsize=(10,10))
plt.plot(np.cumsum(pca.explained_variance_ratio_), 'ro-')
plt.grid()
```

```python
plt.show()
```

```python
pca_new = PCA(n_components=8)
x_new = pca_new.fit_transform(x)
```

```python
from sklearn.cross_validation import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_new, y, test_size = 0.25)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, matthews_corrcoef, classification_report
lr = LogisticRegression()
lr.fit(x_train, y_train)
lr_predict = lr.predict(x_test)
```

```python
lr_conf_matrix = confusion_matrix(y_test, lr_predict)
lr_acc_score = accuracy_score(y_test, lr_predict)
lr_mcc = matthews_corrcoef(y_test, lr_predict)
lr_rep = classification_report(y_test, lr_predict)
print('Confusion Matrix \n', lr_conf_matrix, '\n')
print('Overall Accuracy: ', lr_acc_score*100)
print("Mathews Coeficient Correlation: ", lr_mcc*100)
print('\n', lr_rep)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
dt_predict = dt.predict(x_test)
```

```python
dt_conf_matrix = confusion_matrix(y_test, dt_predict)
dt_acc_score = accuracy_score(y_test, dt_predict)
dt_mcc = matthews_corrcoef(y_test, dt_predict)
dt_rep = classification_report(y_test, dt_predict)
print('Confusion Matrix \n', dt_conf_matrix, '\n')
print('Overall Accuracy: ', dt_acc_score*100)
print("Mathews Coeficient Correlation: ", dt_mcc*100)
print('\n', dt_rep)
```
